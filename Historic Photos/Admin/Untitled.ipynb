```python
import os, re
```

```python
def purge(dir, pattern):
    for f in os.listdir(dir):
        if re.search(pattern, f):
            os.remove(os.path.join(dir, f))
```

```python
def list_dir(dir, pattern):
    for f in os.listdir(dir):
        if re.search(pattern, f):
            print(f)
```

```python
list_dir('./MACRIS/Thumbnail/JPEG',"_[0-9]*_1.jpg")
```
